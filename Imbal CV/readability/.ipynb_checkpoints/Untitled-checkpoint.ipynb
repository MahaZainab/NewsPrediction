```python
import csv
import pandas as pd
import textstat
```

```python
read_file='../data/864newstext.csv'
write_file='../data/864readibility.csv'
rbs=[]
df=pd.read_csv(read_file)
news_into_list=df.values.tolist()
for i in range(len(news_into_list)):
    #news=''
    news=''.join(news_into_list[i])
    #print(textstat.dale_chall_readability_score(news))
    score=textstat.flesch_reading_ease(news)
    #print(score)
    #readability.append(score)
    if score>=90:
        rbs.append(7)
    elif score>=80 and score<=89:
        rbs.append(6)
    elif score >=70 and score<=79:
        rbs.append(5)
    elif score >=60 and score<=69:
        rbs.append(4)
    elif score>=50 and score <=59:
        rbs.append(3)
    elif score>=40 and score<=49:
        rbs.append(2)
    elif score>=30 and score<=39:
        rbs.append(1)
    else:
        rbs.append(0)
    
    
#print(readability)    
    
df2=pd.DataFrame(rbs)

df2.to_csv(write_file)    
```
