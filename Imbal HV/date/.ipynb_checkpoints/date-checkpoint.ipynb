```python
#from datetime import date
from datetime import date
import csv
read_file='3034date.csv'
write_file='date1.csv'
accuracy=[]
dates=open(read_file,'r').read()
dates=dates.splitlines()
for d in dates[1:3035]:
    s_date=d.split('T')
    new_date=s_date[0].split('-')
    year=int(new_date[0])
    month=int(new_date[1])
    day=int(new_date[2])
    date1=date(year, month, day)
    today = date.today()
    delta=today-date1
    accuracy.append(delta.days)
    
print(accuracy)
csv_file=open(write_file, 'w', newline="", encoding='utf-8')
csv_writer=csv.writer(csv_file)
for i in range(len(accuracy)):
    csv_writer.writerow([accuracy[i]])
```
